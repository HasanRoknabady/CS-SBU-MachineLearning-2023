```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/HasanRoknabady/dataset-popularity-/main/OnlineNewsPopularity.csv')

# Display the shape of the dataset
print("Shape of the dataset:", df.shape)

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df.head())

# Display information about the dataset
print("\nInformation about the dataset:")
print(df.info())

# Display summary statistics about the numerical columns
print("\nSummary statistics about the numerical columns:")
print(df.describe())

# Check for missing values
print("\nNumber of missing values in each column:")
print(df.isnull().sum())

# Check for duplicated rows
print("\nNumber of duplicated rows:", df.duplicated().sum())

# Visualize the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=' shares', bins=30)
plt.title("Distribution of shares")
plt.xlabel("Number of shares")
plt.ylabel("Count")
plt.show()

# Visualize the correlation matrix
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation matrix")
plt.show()

# Visualize the pairwise relationships between the numerical variables
sns.pairplot(df[[' n_tokens_title', ' n_tokens_content', ' num_imgs', ' num_videos', ' shares']])
plt.show()

# Visualize the distribution of the categorical variables
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.countplot(x=' data_channel_is_lifestyle', data=df, ax=axs[0, 0])
sns.countplot(x=' data_channel_is_entertainment', data=df, ax=axs[0, 1])
sns.countplot(x=' data_channel_is_bus', data=df, ax=axs[1, 0])
sns.countplot(x=' data_channel_is_tech', data=df, ax=axs[1, 1])
plt.show()
```
