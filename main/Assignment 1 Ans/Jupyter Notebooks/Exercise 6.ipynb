```python
import numpy as np
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes()

# Extract the feature matrix X and the target variable y
X = diabetes.data
y = diabetes.target.reshape(-1, 1)

# Add a column of ones to X for the bias term
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Compute the normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the values of the parameters
print("Parameter values:")
print(theta_best)

# Use the model to make predictions on a new dataset
X_new = np.array([[0.02, 0.05, -0.06, 0.03, 0.01, 0.027, -0.04, 0.031, 0.01, -0.03]])
X_new_b = np.c_[np.ones((1, 1)), X_new]
y_pred = X_new_b.dot(theta_best)
print("Prediction on new data:")
print(y_pred)
```

```python
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()

# Extract the feature matrix X and the target variable y
X = diabetes.data
y = diabetes.target.reshape(-1, 1)

# Split the data into training and testing sets
X_train = X[:-20]
X_test = X[-20:]
y_train = y[:-20]
y_test = y[-20:]

# Add a column of ones to X for the bias term
X_train_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]

# Compute the normal equation
theta_best = np.linalg.inv(X_train_b.T.dot(X_train_b)).dot(X_train_b.T).dot(y_train)

# Print the values of the parameters
print("Parameter values:")
print(theta_best)

# Make predictions on the testing set
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
y_pred = X_test_b.dot(theta_best)

# Compute the mean squared error
mse = np.mean((y_pred - y_test) ** 2)
print("Mean squared error:", mse)

# Plot the predictions against the actual values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual values")
plt.ylabel("Predictions")
plt.show()
```

```python
# Extract features and target
X = diabetes.data
y = diabetes.target

# Add bias term to X
X = np.c_[X, np.ones(X.shape[0])]

# Compute the normal equation
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Compute predictions on the training set
y_pred = X.dot(theta)

# Compute the mean squared error (MSE) and the coefficient of determination (R^2)
MSE = np.mean((y_pred - y) ** 2)
R2 = 1 - np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2)

# Plot the predictions vs. the actual values
plt.scatter(y, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Predictions vs. Actual (MSE = {:.2f}, R2 = {:.2f})'.format(MSE, R2))
plt.show()
```
